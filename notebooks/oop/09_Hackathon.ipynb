## 1. Make a card game with a Deck of cards

In this hackathon you will be extending the already existing deck of cards below to create a game.

The game goes like this:
- 2+ players in a game
- Each player gets one full deck of cards, shuffled
- Each player deals 15 cards and discards them
- With the remaining cards, the players add up the total number of face cards (J, Q, K)
- The player with the most face cards wins
- If there is a draw, the player with the most Ks wins (then most Qs, Js etc.)

The goal of this challenge is to use the `__lt__`, `__gt__`, `__eq__`, `__ge__`, `__le__` dunder methods to compare the decks with the remaining cards.

```python
import collections
from random import shuffle

Card = collections.namedtuple('Card', ['rank', 'suit'])

class Deck:
    ranks = '23456789TJQKA'
    suits = '♠♥♦♣'
    
    def __init__(self, name):
        self._original_cards = [
            Card(rank, suit)
            for suit in self.suits
            for rank in self.ranks
        ]
        self._name = name
   
    def __len__(self):
        return len(self._original_cards)
        
    def __str__(self):
        return f'Deck(suits={self.suits}, ranks={self.ranks})'
    
    def __getitem__(self, position):
        return self._original_cards[position]
    
    def __setitem__(self, ind, value):
        self._original_cards[ind] = value

    def __lt__(self, other):
        return self.num_jqk < other.num_jqk
    
    def __eq__(self, other):
        return self.num_jqk == other.num_jqk
    
    def __gt__(self, other):
        return self.num_jqk > other.num_jqk
    
    def deal(self):
        return self._original_cards.pop()
    
    def check_ace(self):
        return self._original_cards[-1].rank == 'A'
    
    def win(self, other):
        print(self._name, 'has', self.num_jqk, 'face cards of which', self.num_k, 'K,', self.num_q, 'Q,', self.num_j, 'J.')
        print(other._name, 'has', other.num_jqk, 'face cards of which', other.num_k, 'K,', other.num_q, 'Q,', other.num_j, 'J.')
        if self < other:
            return self._name + " loses to " + other._name
        elif self > other:
            return self._name + " wins from " + other._name
        else:
            tie_msg = "Both players have the same number of face cards"
            tie_breaked = 0
            for letter, attr in zip(['K', 'Q', 'J'],['num_k', 'num_q', 'num_j']):
                if tie_breaked == 0:
                    num_card_self = getattr(self, attr)
                    num_card_other = getattr(other, attr)
                    if num_card_self != num_card_other:
                        tie_breaked = 1
                        if num_card_self > num_card_other:
                            tie_msg += " but " + self._name + " wins with the most " + letter
                        else:
                            tie_msg += " but " + other._name + " wins with the most " + letter
            if tie_breaked == 0:
                tie_msg += " and of each letter individually"
            return tie_msg
    
    @property
    def num_q(self):
        return len([card for card in self._original_cards if card.rank == "Q"])
    
    @property
    def num_j(self):
        return len([card for card in self._original_cards if card.rank == "J"])
    
    @property
    def num_k(self):
        return len([card for card in self._original_cards if card.rank == "K"])
    
    @property
    def num_jqk(self):
        return self.num_q + self.num_j + self.num_k
    
    @property
    def cards(self):
        return self._original_cards

# Create the decks        
deck1 = Deck("Player A")
deck2 = Deck("Player B")

# Shuffle the decks
shuffle(deck1)
shuffle(deck2)

# Draw 15 cards of each deck
for _ in range(15):
    deck1.deal()
    deck2.deal()

# Check for win/draw
deck1.win(deck2)
```

## 2. Make a bank account

Using what you've learned about object oriented programming, make a class that works as a bank account.

Your banking account should have the following properties (and any other functionality you can think of):
- When opened (initialized) it should have a starting balance of 0 
- The owner should be able to 
    - deposit money (add money to the balance)
    - withdraw money, as long as the withdraw amount doesn't put them over their overdraft limit
    - increase their overdraft limit (starts at 0) by 100 each time

```python
class BankAccount:
    
    def __init__(self, name, balance = 0, overdraft = 0):
        self.name = name
        self.balance = balance
        self.overdraft = overdraft
    
    def __str__(self):
        return "The bankaccount of {}".format(self.name)

account = BankAccount("christa")
print(account)
```
